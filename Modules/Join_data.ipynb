```python
from PIL import Image
import numpy as np

mask_image = Image.open('/content/MASK_SEGMENT.png').convert('L')  # Convertir a escala de grises
depth_image = Image.open('/content/colored_depth_image.png')

mask_array = np.array(mask_image)
depth_array = np.array(depth_image)

if mask_array.shape != depth_array.shape[:2]:
    mask_image = mask_image.resize(depth_image.size, Image.NEAREST)
    mask_array = np.array(mask_image)

boolean_mask = mask_array == 255

masked_depth = np.zeros_like(depth_array)
masked_depth[boolean_mask] = depth_array[boolean_mask]

result_image = Image.fromarray(masked_depth)
result_image.save('/content/height_buildings.png')

```
